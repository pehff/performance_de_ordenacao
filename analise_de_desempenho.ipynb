### Função Bubble sort

Bubble sort é o algoritmo mais simples, mas o menos eficientes. Neste algoritmo cada elemento da posição i será comparado com o elemento da posição i + 1, ou seja, um elemento da posição 2 será comparado com o elemento da posição 3. Caso o elemento da posição 2 for maior que o da posição 3, eles trocam de lugar e assim sucessivamente. Por causa dessa forma de execução, o vetor terá que ser percorrido quantas vezes que for necessária, tornando o algoritmo ineficiente para listas muito grandes.

![Bubble Sort](https://arquivo.devmedia.com.br/artigos/Bruno_Almeida_Honorato/Algoritmos-Resolucao-Problemas/2.jpg)

<ul>
    <li>É verificado se o 3 é maior que 5, por essa condição ser falsa, não há troca.</li>
    <li>É verificado se o 5 é maior que 1, por essa condição ser verdadeira, há uma troca.</li>
    <li>É verificado se o 5 é maior que 2, por essa condição ser verdadeira, há uma troca.</li>
    <li>É verificado se o 5 é maior que 4, por essa condição ser verdadeira, há uma troca.</li>
    <li>O método retorna ao início do vetor realizando os mesmos processos de comparações, isso é feito até que o vetor esteja ordenado.</li>
</ul>

```python
def bubble_sort(lista):
    elementos = len(lista)-1
    ordenado = False
    while not ordenado:
        ordenado = True
        for i in range(elementos):
            if lista[i] > lista[i+1]:
                lista[i], lista[i+1] = lista[i+1],lista[i]
                ordenado = False
    return lista
```

### Função Selection Sort

Este algoritmo é baseado em se passar sempre o menor valor do vetor para a primeira posição (ou o maior dependendo da ordem requerida), depois o segundo menor valor para a segunda posição e assim sucessivamente, até os últimos dois elementos.
Neste algoritmo de ordenação é escolhido um número a partir do primeiro, este número escolhido é comparado com os números a partir da sua direita, quando encontrado um número menor, o número escolhido ocupa a posição do menor número encontrado. Este número encontrado será o próximo número escolhido, caso não for encontrado nenhum número menor que este escolhido, ele é colocado na posição do primeiro número escolhido, e o próximo número à sua direita vai ser o escolhido para fazer as comparações. É repetido esse processo até que a lista esteja ordenada.

![Selection Sort](https://arquivo.devmedia.com.br/artigos/Bruno_Almeida_Honorato/Algoritmos-Resolucao-Problemas/3.jpg)

<ul>
    <li>Neste passo o primeiro número escolhido foi o 3, ele foi comparado com todos os números à sua direita e o menor número encontrado foi o 1, então os dois trocam de lugar.</li>
    <li>O mesmo processo do passo 1 acontece, o número escolhido foi o 5 e o menor número encontrado foi o 2.</li>
    <li>Não foi encontrado nenhum número menor que 3, então ele fica na mesma posição.</li>
    <li>O número 5 foi escolhido novamente e o único número menor que ele à sua direita é o 4, então eles trocam.
Vetor já ordenado.</li>
</ul>






```python
def selection_sort(lista):
    for i in range(len(lista)-1,0,-1):
        posicao_max =0
        for j in range(1,i+1):
            if lista[j]> lista[posicao_max]:
                posicao_max = j    
        temp = lista[i]
        lista[i] = lista[posicao_max]
        lista[posicao_max] = temp
    return lista
```

### Função Insertion Sort

O Insertion sort é um algoritmo simples e eficiente quando aplicado em pequenas listas. Neste algoritmo a lista é percorrida da esquerda para a direita, à medida que avança vai deixando os elementos mais à esquerda ordenados.
O algoritmo funciona da mesma forma que as pessoas usam para ordenar cartas em um jogo de baralho como o pôquer.

![Insertion Sort](https://arquivo.devmedia.com.br/artigos/Bruno_Almeida_Honorato/Algoritmos-Resolucao-Problemas/4.jpg)

<ul>
    <li>Neste passo é verificado se o 5 é menor que o 3, como essa condição é falsa, então não há troca.</li>
    <li>É verificado se o quatro é menor que o 5 e o 3, ele só é menor que o 5, então os dois trocam de posição.</li>
    <li>É verificado se o 2 é menor que o 5, 4 e o 3, como ele é menor que 3, então o 5 passa a ocupar a posição do 2, o 4 ocupa a posição do 5 e o 3 ocupa a posição do 4, assim a posição do 3 fica vazia e o 2 passa para essa posição.</li>
    <li>O mesmo processo de comparação acontece com o número 1, após esse processo o vetor fica ordenado.</li>
</ul>

```python
def insertion_sort(lista):
    for i in range(1, len(lista)):
        valor_atual = lista[i]
        posicao = i

        while (posicao > 0) and (lista[posicao-1] > valor_atual):
            lista[posicao] = lista[posicao-1]
            posicao -= 1

        lista[posicao] = valor_atual
    return lista
```

### Função Quick Sort

O Quicksort é o algoritmo mais eficiente na ordenação por comparação. Nele se escolhe um elemento chamado de pivô, a partir disto é organizada a lista para que todos os números anteriores a ele sejam menores que ele, e todos os números posteriores a ele sejam maiores que ele. Ao final desse processo o número pivô já está em sua posição final. Os dois grupos desordenados recursivamente sofreram o mesmo processo até que a lista esteja ordenada.

![Quick Sort](https://arquivo.devmedia.com.br/artigos/Bruno_Almeida_Honorato/Algoritmos-Resolucao-Problemas/5.jpg)

<ul>
    <li>O número 3 foi escolhido como pivô, nesse passo é procurado à sua direita um número menor que ele para ser passado para a sua esquerda. O primeiro número menor encontrado foi o 1, então eles trocam de lugar.</li>
    <li>Agora é procurado um número à sua esquerda que seja maior que ele, o primeiro número maior encontrado foi o 5, portanto eles trocam de lugar.</li>
    <li>O mesmo processo do passo 1 acontece, o número 2 foi o menor número encontrado, eles trocam de lugar.</li>
    <li>O mesmo processo do passo 2 acontece, o número 4 é o maior número encontrado, eles trocam de lugar.</li>
    <li>O vetor desse exemplo é um vetor pequeno, portanto ele já foi ordenado, mas se fosse um vetor grande, ele seria dividido e recursivamente aconteceria o mesmo processo de escolha de um pivô e comparações.</li>
</ul>

```python
def quick_sort(lista):
    qsort(lista, 0, len(lista)-1)
    return lista

def qsort(lista, inicio, fim):
    if inicio < fim:
        p = partition(lista, inicio, fim)
        qsort(lista, inicio, p)
        qsort(lista, p + 1, fim)

def partition(lista, inicio, fim):
    pivot = lista[inicio]
    i, j = inicio-1, fim+1
    while True:
        i += 1
        j -= 1
        while(lista[i] < pivot): i+= 1
        while(lista[j] > pivot ): j-= 1

        if i >= j:
            return j

        lista[i], lista[j] = lista[j], lista[i]
```

### Definição Merge Sort

O merge sort, ou ordenação por mistura, é um exemplo de algoritmo de ordenação por comparação do tipo dividir-para-conquistar.
Sua ideia básica consiste em Dividir (o problema em vários subproblemas e resolver esses subproblemas através da recursividade) e Conquistar (após todos os subproblemas terem sido resolvidos ocorre a conquista que é a união das resoluções dos subproblemas). Como o algoritmo Merge Sort usa a recursividade, há um alto consumo de memória e tempo de execução, tornando esta técnica não muito eficiente em alguns problemas.

![Merge Sort](https://upload.wikimedia.org/wikipedia/commons/c/cc/Merge-sort-example-300px.gif)

```python
def merge_sort(lista):
    if len(lista)>1:
        meio = len(lista)//2
        parte_esquerda = lista[:meio]
        parte_direita = lista[meio:]

        merge_sort(parte_esquerda)
        merge_sort(parte_direita)

        i=0
        j=0
        k=0
        
        while i < len(parte_esquerda) and j < len(parte_direita):
            if parte_esquerda[i] < parte_direita[j]:
                lista[k] = parte_esquerda[i]
                i=i+1
            else:
                lista[k] = parte_direita[j]
                j=j+1
            k=k+1  

        while i < len(parte_esquerda):
            lista[k] = parte_esquerda[i]
            i=i+1
            k=k+1
            
        while j < len(parte_direita):
            lista[k] = parte_direita[j]
            j=j+1
            k=k+1
            
    return lista
```

### Definição Shell Sort

Shell sort é o mais eficiente algoritmo de classificação dentre os de complexidade quadrática. É um refinamento do método de inserção direta. O algoritmo difere do método de inserção direta pelo fato de no lugar de considerar o array a ser ordenado como um único segmento, ele considera vários segmentos sendo aplicado o método de inserção direta em cada um deles. Basicamente o algoritmo passa várias vezes pela lista dividindo o grupo maior em menores.

![Shell Sort](https://www.tutorialspoint.com/data_structures_algorithms/images/shell_sort.jpg)

```python
def shell_sort(lista):
    h = 1
    n = len(lista)
    while h > 0:
            for i in range(h, n):
                c = lista[i]
                j = i
                while j >= h and c < lista[j - h]:
                    lista[j] = lista[j - h]
                    j = j - h
                    lista[j] = c
            h = int(h / 2.2)
    return lista
```

### Definição Heap Sort

O heapsort utiliza uma estrutura de dados chamada heap, para ordenar os elementos à medida que os insere na estrutura. Assim, ao final das inserções, os elementos podem ser sucessivamente removidos da raiz da heap, na ordem desejada, lembrando-se sempre de manter a propriedade de max-heap.
A heap pode ser representada como uma árvore (uma árvore binária com propriedades especiais) ou como um vetor. Para uma ordenação decrescente, deve ser construída uma heap mínima (o menor elemento fica na raiz). Para uma ordenação crescente, deve ser construído uma heap máxima (o maior elemento fica na raiz).

![Heap Sort](https://upload.wikimedia.org/wikipedia/commons/4/4d/Heapsort-example.gif)

```python
def agrupar(lista, n, i):
    maior = i
    l = 2 * i + 1
    r = 2 * i + 2 
 
    if l < n and lista[i] < lista[l]:
        maior = l
 
    if r < n and lista[maior] < lista[r]:
        maior = r
 
    if maior != i:
        lista[i],lista[maior] = lista[maior],lista[i]
        agrupar(lista, n, maior)

def heap_sort(lista):
    n = len(lista)
 
    for i in range(n, 0, -1):
        agrupar(lista, n, i-1)
 
    
    for i in range(n-1, 0, -1):
        lista[i], lista[0] = lista[0], lista[i]  
        
        agrupar(lista, i, 0)
    
    return lista
```

### Definição Busca Linear

```python
def busca_linear(num, lista):
    tentativas = 0
    for i in lista:
        tentativas += 1
        if num == i:
            return (tentativas - 1)
            break
```

### Definição Busca Binária

```python
def busca_binaria(lista, num):
    menor = 0
    maior = len(lista) - 1 
    tentativas = 1
    
    while menor <= maior:
        meio = (menor + maior) // 2
        
        if num < lista[meio]:
            maior = meio - 1
        elif num > lista[meio]:
            menor = meio + 1  
        else:
            return tentativas
            break
        
        tentativas += 1
    
    return "Número não encontrado"
```

## Cálculo da média do tempo de execução das ordenações

#### Bubble Sort

```python
def media_ord_bubble(num_execucoes, vetor):
    soma = 0
    
    for i in range(0, num_execucoes):
        lista = copy.deepcopy(vetor)
        
        inicio_ordenacao = time.time()
        bubble_sort(lista)
        fim_ordenacao = time.time()
        
        tempo_ordenacao = fim_ordenacao - inicio_ordenacao
        #print('{}ª execução: {} s'.format(i+1, tempo_ordenacao))
        soma += tempo_ordenacao
    
    media = soma / num_execucoes
    
    return media
```

## Cálculo da média do tempo de execução das buscas

#### Busca Linear Ordenada

```python
def media_busca_linear_ord(num_execucoes, vetor, num):
    soma = 0
    
    for i in range(0, num_execucoes):
        
        lista = copy.deepcopy(vetor)
        quick_sort(lista)
        
        inicio_busca_linear_ord = time.time()
        busca_linear(num, lista)
        fim_busca_linear_ord = time.time()
        
        tempo_busca_ord = fim_busca_linear_ord - inicio_busca_linear_ord
        #print('Tempo de execução da busca linear com a lista ordenada: {}'.format(tempo_busca_ord))
        
        soma += tempo_busca_ord
    
    media = soma / num_execucoes
    
    return media
```

```python
def media_busca_binaria(num_execucoes, vetor, num):
    soma = 0
    
    for i in range(0, num_execucoes):
        lista = copy.deepcopy(vetor)
        quick_sort(lista)  
        
        inicio_busca_binaria = time.time() 
        busca_binaria(lista, num)
        fim_busca_binaria = time.time()
        
        tempo_busca_binaria = fim_busca_binaria - inicio_busca_binaria
        #print('Tempo de execução da busca binária: {}'.format(tempo_busca_binaria))
        
        soma += tempo_busca_binaria
    
    media = soma / num_execucoes
    
    return media
```

```python
def gerar_grafico_comparacao_busca(execucoes, vetor, randomico):
    
    if randomico == "sim":
        numero_in = randint(0,15000)
        print("Valor gerado automaticamente: {}".format(numero_in))
    else:
        numero_in = int(input('Qual número você deseja pesquisar: '))
    
    busca = ('Linear Ord.', 'Linear Não Ord.', 'Binária')
    tempo = (media_busca_linear_ord(execucoes, vetor, numero_in),
             media_busca_linear_nao_ord(execucoes, vetor, numero_in),
             media_busca_binaria(execucoes, vetor, numero_in))
    
    lista = copy.deepcopy(vetor)
    quick_sort(lista) 
    print("\nResultado da busca binária: ",busca_binaria(lista, numero_in))

    largura = 0.5
    
    sns.set(style="whitegrid", context="talk")
    plt.xlabel('Buscas')
    plt.ylabel('Tempo de execução (s)')
    plt.title('Gráfico de execução dos algoritmos de busca')
    
    plt.bar(busca, tempo, largura)
    plt.show()
    
    tabela = pd.Series(tempo, busca)
    print(tabela)
```
